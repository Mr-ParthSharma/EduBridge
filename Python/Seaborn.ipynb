```python
sns.set_style("dark")
sns.kdeplot(cars['mpg'])
plt.show()
```

```python
sns.set_style("dark")
sns.kdeplot(cars['mpg'], color ='yellow', shade=True)
plt.show()
```

```python
sns.set_style("dark")
sns.kdeplot(cars['mpg'], color ='yellow',vertical=True, shade=True)
plt.show()
```

```python
sns.set_style("dark")
sns.kdeplot(cars['mpg'],cars['horsepower'], color ='green', shade=True)
plt.show()
```

```python
sns.set_style('whitegrid')
cyl_4 = cars[cars.cylinders==4]
cyl_8 = cars[cars.cylinders==8]

plt.figure(figsize=(8,6))

sns.kdeplot(cyl_4.horsepower, cyl_4.mpg, cmap = 'Blues', shade=True, shade_lowest=False)
sns.kdeplot(cyl_8.horsepower, cyl_8.mpg, cmap = 'Reds', shade=True, shade_lowest=False)

plt.xlabel('Hoesepower')
plt.ylabel('miles per gallon (MPG)')

plt.annotate("4 cylinders",(105,32),color='b',fontsize=16,fontweight='bold')
plt.annotate("8 cylinders",(190,18),color='r',fontsize=16,fontweight='bold')
plt.show()
```
